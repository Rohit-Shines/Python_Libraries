# Advacned filtering in pandas



```python
# filtering data with all conditions to satisfy
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') | (df['HP'] > 70)]
```

```python
# reset index
df.reset_index(drop=True, inplace=True)
df
```

# Contains function

```python
# filter the names with Mega in an specific column ex - appleMega , catMeagapple
df.loc[df['Name'].str.contains('Mega')]
```

```python
# ~ works as not. it will print reverse of that codition
df.loc[~df['Name'].str.contains('Mega')]
```

# Regular expression 

```python
import re # regular expresion

# this will search for either Mega or aur (Mega|aur) in column Name
df.loc[~df['Name'].str.contains('Mega|aur', regex=True)]
```

```python
#  Search pikahu with pik
df.loc[df['Name'].str.contains('Pik')]
```

```python
#change specific column or row details with other name like cntrlh replace
df.loc[df['Type 1']=='Fire','Type 1']='Flamer'
df
```
